```python
# Importando o Pandas
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Ler o arquivo de base/exemplo
df = pd.read_csv('Dados_Entrada/SNIS_Florianopolis_Residuos_Limpo.csv')
df
```

```python
# Método Describe - Vou aplicar para a coluna CO119 que dá a quantidade total de RDO e RPU coletada por todos os tipos de prestadores de serviço. 
dados = df['CO119 - Quantidade total de RDO e RPU coletada por todos os agentes'].describe()
```

```python
print(dados)
```

```python
# (a) Qual a média da sua série de dados?
media = dados['mean']
```

```python
print(f"A média de RDO e RPU coletada por todos os agentes dos anos de 2002 até 2022 é de: {media}")
```

```python
# (b) Qual o valor do meio (percentil 50%)?
mediana = dados['50%']
```

```python
print(f"A Mediana de RDO e RPU coletada por todos os agentes dos anos de 2002 até 2022 é de: {mediana}")
```

```python
# (c) Qual a faixa de valores entre os percentis 25% e 75%?
percentil_25 = dados['25%']
percentil_75 = dados['75%']
```

```python
print(f"Faixa entre percentil 25% e 75% para os valores de RDO e RPU coletada por todos os agentes dos anos de 2002 até 2022 é de: {percentil_25} - {percentil_75}")
```

```python
# Plotando os dados
plt.figure(figsize=(10, 6))
plt.plot(df['Ano de Referência'], df['CO119 - Quantidade total de RDO e RPU coletada por todos os agentes'], label='Quantidade Total Coletada', color='blue')
```

```python
# Plotando os dados
plt.figure(figsize=(10, 6))
plt.plot(df['Ano de Referência'], df['CO119 - Quantidade total de RDO e RPU coletada por todos os agentes'], label='Quantidade Total Coletada', color='blue')
# Plotando a média
plt.axhline(media, color='green', linestyle='--', label=f'Média: {media:.2f}')
```

```python
# Extraindo a média e o desvio padrão do describe
media = dados['mean']  # Média calculada
desvio_padrao = dados['std']  # Desvio padrão calculado
```

```python
# Plotando os dados
plt.figure(figsize=(10, 6))
plt.plot(df['Ano de Referência'], df['CO119 - Quantidade total de RDO e RPU coletada por todos os agentes'], label='Quantidade Total Coletada', color='blue')

# Ajustar o eixo X para mostrar os anos inteiros
plt.xticks(df['Ano de Referência'], rotation=45)
           
# Plotando a linha da média
plt.axhline(media, color='green', linestyle='--', label=f'Média: {media:.2f}')

# Plotando a área do desvio padrão
plt.fill_between(df['Ano de Referência'], media - desvio_padrao, media + desvio_padrao, color='gray', alpha=0.2, label=f'Desvio Padrão: ±{desvio_padrao:.2f}')

# Adicionando título, legendas e rótulos
plt.title('Gráfico de Linha - Quantidade Total de RDO e RPU Coletada por Todos os Agentes')
plt.xlabel('Ano de Referência')
plt.ylabel('Quantidade de RDO e RPU Coletada (toneladas)')
plt.legend()

# Adicionar linhas secundárias horizontais com menor opacidade
plt.grid(True, which='both', axis='y', linestyle='--', color='gray', alpha=0.5) 
```
